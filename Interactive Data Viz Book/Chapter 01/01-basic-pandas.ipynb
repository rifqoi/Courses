# Categorical Data

cut, color, clarity adalah data kategorikal \
carat, depth, table, price, x, y, z adalah data numerikal

Kolom cut, color, clarity adalah data kategorikal yang bersifat ordinal \
Data ordinal adalah data kategorikal yang bisa diurutkan berdasarkan rankingnya atau peringkatnya\
misal: Pangkat, Ranking Tim Liga Satu


```python
print(diamonds_df.cut.nunique(),'\n', diamonds_df.cut.unique(), '\n')
print(diamonds_df.color.nunique(), '\n', diamonds_df.color.unique(), '\n')
print(diamonds_df.clarity.nunique(),'\n', diamonds_df.clarity.unique())
```

Dapat dilihat kolom cut menggambarkan ranking dari diamond \
Data color mengurutkan warna dari diamond dari J (worst) - D (Best)\
Data clarity mengurutkan kejernihan dari diamond I1
(worst), SI1, SI2, VS1, VS2, VVS1, VVS2, and IF (best).

Untuk menghitung mean, median, dan mode dari data kategorikal kita tambahkan parameter\
`include='category'` pada method describe()

# Add columns to dataframe

```python
# Dua cara untuk melihat Dataframe dengan column cut == "Ideal"
diamonds_low_df = diamonds_df.loc[diamonds_df['cut'] == 'Ideal']
diamonds_low_df = diamonds_df[diamonds_df['cut'] == 'Ideal']

diamonds_low_df
```

Menambah column pada dataframe

```python
diamonds_df['price_per_carat'] = diamonds_df['price'] / diamonds_df['carat']
diamonds_df
```

Pakai np.where(<condition>) untuk mengklasifikasi price_per_carat yang mahal

```python
diamonds_df['price_per_carat_is_high'] = np.where(diamonds_df['price_per_carat'] > 3500, 1, 0)
diamonds_df
```

# Applying Functions on DataFrame Column

```python
import math
# Ceil harga price menggunakan apply
diamonds_df['new_price'] = diamonds_df.price * 1.3
diamonds_df['new_price'] = diamonds_df.new_price.apply(math.ceil)

# Menggunakan lambda untuk ceil ke nilai kelipatan 100 terdekat
diamonds_df.new_price = diamonds_df.new_price.apply(lambda x: math.ceil(x/100)*100)
diamonds_df
```

# Applying Functions on Multiple Columns

```python
# Menggunakan apply pada multiple columns
# 'yes' apabila cut = Ideal dan color = D
def is_desired(df):
    return 'yes' if df['cut'] == 'Ideal' and df['color'] == 'D' else 'no'

# parameter axis=1 untuk memilih column
diamonds_df['is_desired'] = diamonds_df.apply(is_desired, axis=1)
diamonds_df
```

# Delete columns on DataFrame

```python
# Menggunakan drop dan parameter columns
diamonds_df.drop(columns=['new_price'])

# Menggunakan parameter inplace untuk menghapus column tanpa reassign ke variable
diamonds_df.drop(columns=['new_price'], inplace=True)
```
